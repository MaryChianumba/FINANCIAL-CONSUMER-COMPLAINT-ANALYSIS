CHECKING FOR SEASONAL PATTERN IN CONSUMER COMPLAINTS

```python
df['Date submitted'] = pd.to_datetime(df['Date submitted'])
df['Month'] = df['Date submitted'].dt.month

monthly_complaints = df.groupby('Month').size()

plt.figure(figsize=(10, 6))
monthly_complaints.plot(kind='line', marker='o')
plt.title('Monthly Complaints Over Time')
plt.xlabel('Month')
plt.ylabel('Number of Complaints')
plt.xticks(range(1, 13), ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(True)
plt.show()
```

From the above illustration, consumer complaints were high in the month of july

PRDOUCTS WITH THE MOST COMPLAINT AND COMMON ISSUES

```python
df['Product'].unique()
```

```python
# Count complaints by product
product_counts = df['Product'].value_counts()

print(product_counts)
```

SHORTENING THE PRODUCT NAMES FOR EFFECTIVE PLOTTING

```python
abbreviation_map = {
    'Checking or savings account': 'Checking/Savings',
    'Credit card or prepaid card': 'Credit Card/Prepaid',
    'Credit reporting, credit repair services, or other personal consumer reports': 'Credit Reporting',
    'Mortgage': 'Mortgage',
    'Money transfer, virtual currency, or money service': 'Money Transfer',
    'Debt collection': 'Debt Collection',
    'Vehicle loan or lease': 'Vehicle Loan/Lease',
    'Payday loan, title loan, or personal loan': 'Payday/Title/Personal Loan',
    'Student loan': 'Student Loan'}

# Replace the product names with their shortened versions
df['Product_short'] = df['Product'].replace(abbreviation_map)
```

```python
# Plotting the number of complaints by product
df['Product_short'].value_counts().plot(kind='bar')
plt.title('Complaints by Product')
plt.xlabel('Product')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=90)
plt.show()
```

MOST COMMON ISSUES WITHIN THE MOST COMMON COMPLAINTS

```python
# Calculating the count of complaints for each product
product_counts = df['Product_short'].value_counts()

# Identifying the product(s) with the highest count of complaints
most_complained_products = product_counts[product_counts == product_counts.max()].index

# For each of the most complained product(s), finding the most common issue(s)
for product in most_complained_products:
    # Filtering the DataFrame for the current most complained product
    product_df = df[df['Product_short'] == product]
    
    # Getting the most common issues and their counts for the current most complained product
    common_issues = product_df['Issue'].value_counts()
    
    # Plotting the bar chart for the most common issues
    plt.figure(figsize=(10, 6))
    common_issues.plot(kind='bar')
    plt.title(f'Most Common Issues for {product}')
    plt.xlabel('Issue')
    plt.ylabel('Number of Complaints')
    plt.xticks(rotation=90)
    plt.tight_layout()
    plt.show()
```

HOW COMPLAINTS WERE TYPICALLY RESOLVED

```python
# Count complaints by company response
response_counts = df['Company response to consumer'].value_counts()

# Plotting the distribution of company responses
plt.figure(figsize=(10, 6))
response_counts.plot(kind='bar')
plt.title('Distribution of Company Responses')
plt.xlabel('Company Response')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=90)
plt.show()
```

Most of the complaints are typically resolved through closed explanation 

TIMELY RESPONSE OF COMPLAINTS

```python
# Calculating the count of timely and untimely responses
response_counts = df['Timely response?'].value_counts()

# Plotting the bar chart
plt.figure(figsize=(8, 6)) 
response_counts.plot(kind='bar', color=['green', 'red'])
plt.title('Timely Responses')
plt.xlabel('Timely Response')
plt.ylabel('Number of Complaints')
plt.xticks(rotation=0)  
plt.tight_layout()  
plt.show()
```

Complaints were mostly resolved on time

```python
# Filtering the DataFrame for complaints with untimely responses
untimely_responses = df[df['Timely response?'] == 'No']

# Grouping by issue and count the occurrences of each issue
issue_counts = untimely_responses['Issue'].value_counts()

# Printing the top 10 most common issues
print("Top 10 most common issues in complaints with untimely responses:")
print(issue_counts.head(10))
```

The most common issue in complaints with untimely response is managing an account
